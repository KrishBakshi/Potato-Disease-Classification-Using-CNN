```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image_dataset_from_directory
```

```python
df_path = './Dataset'
```

## Creating Image Data Pipeline

```python
import os

dim1 = []
dim2 = []
for folders in os.scandir(df_path):
    for images in os.scandir(folders.path):
#         print(images)
        img = plt.imread(images.path)
        d1,d2,c = img.shape
        dim1.append(d1)
        dim2.append(d2)        
```

```python
np.mean(dim1)
```

```python
np.mean(dim2)
```

```python
IMG_SIZE = 256
BATCH_SIZE = 32
CHANNEL = 3
```

```python
df = image_dataset_from_directory(df_path,
                                 seed=123,
                                 image_size=(IMG_SIZE,IMG_SIZE),
                                 batch_size=BATCH_SIZE)
```

## EDA

```python
classes = df.class_names
```

```python
classes
```

```python
plt.figure(figsize=(10,10))
for image_batch,labels_batch in df.take(2):
    for i in range(12):
        plt.subplot(3,4,i+1)
        plt.imshow(image_batch[i].numpy().astype("uint8"))
        plt.title(classes[labels_batch[i]])
        plt.axis('off')
```
